```python
import pandas as pd
#import kaleido
#import plotly.io as pio
import plotly.express as px
```

```python
trains = pd.read_csv('../data/train_final.csv')
airplanes = pd.read_csv('../data/airplanes_final.csv')
```

```python
trains.dropna(inplace=True)
airplanes.dropna(inplace=True)
```

```python
trains.rename(columns={'CASINJRR' : "Injured", 'CASKLDRR': "Killed"}, inplace=True)
trains.sort_values(by='DATE', inplace=True)

train_fig = px.line(trains, x='DATE', y=['Injured', 'Killed'], color_discrete_sequence=["lightgreen", "darkgreen"],
              labels={
                "value" : "Number of People",
                "DATE" : "Date",
                "variable" : "Measurement"
              }, 
              title="Railroad Deaths and Injuries over Time")
train_fig.update_xaxes(
    rangeslider_visible=True)
train_fig.show()
```

```python
dir_path = "./img/line_chart/"
img_type = "Railroad"
train_fig.write_image(f"{dir_path}/{img_type}.png")
```

```python
airplanes.sort_values(by='Date', inplace=True)
airplanes.rename(columns={"Fatalities": "Onboard"}, inplace=True)

plane_fig = px.line(airplanes, x='Date', y=['Onboard', 'Ground'], color_discrete_sequence=["deepskyblue", "darkblue"],
              labels={
                "value" : "Number of People",
                "DATE" : "Date",
                "variable" : "Fatalities"
              },
              title="Plane Onboard and Ground Deaths over Time")
plane_fig.update_xaxes(
    rangeslider_visible=True)
plane_fig.show()
```

```python
img_type = "Plane"
train_fig.write_image(f"{dir_path}/{img_type}.png")
```

```python
concat = pd.concat([trains, airplanes])
concat.sort_values(by='Date', inplace=True)
fig = px.line(concat, x='Date', y=['Onboard', 'Ground', 'Injured', 'Killed'], color_discrete_sequence=["deepskyblue", "darkblue", "lightgreen", "darkgreen"],
              labels={
                "value" : "Number of People",
                "DATE" : "Date",
                "variable" : "Fatalities"
              },
              title="Number of People Killed In Plane Crashes over Time")
fig.update_xaxes(
    rangeslider_visible=True)
fig.show()
```
