```python
import pandas as pd

trains = pd.read_csv('../data/train_final.csv')
airplanes = pd.read_csv('../data/airplanes_final.csv')
```

```python
trains.dropna(inplace=True)
airplanes.dropna(inplace=True)
```

```python
#trains['YEAR'] = trains['DATE'].str[:4]
from datetime import datetime
trains['DATETIME'] = pd.to_datetime(trains['DATE'])
trains['YEAR'] = trains['DATETIME'].dt.year
trains['YEAR']
```

```python
airplanes['Datetime'] = pd.to_datetime(airplanes['Date'])
airplanes['Year'] = airplanes['Datetime'].dt.year
airplanes['Year']
```

```python
# INITIALIZE GRAPH OBJECT
import plotly.graph_objects as go
fig = go.Figure()


#AVERAGE OUT THE YEARS THAT ARE THE SAME??
train_avg = trains.groupby('YEAR', as_index=False)[('TRNSPD', 'TONS')].mean()
print(train_avg)
print(type(train_avg))

train_num = 0
all_years = []
sorted_train_years = train_avg['YEAR'].sort_values(ascending=True,inplace=False)
for year in sorted_train_years: 
    all_years.append(year)
    train_num += 1  
    df = train_avg[train_avg['YEAR'] == year]  
    fig.add_trace(go.Scatter(
        x=df['TRNSPD'],
        y=df['TONS'],
        mode="lines",
        marker=dict(
            color="green",
            showscale=True,
            symbol="circle"
        ),
        visible=False,
        name="TRNSPD vs TONS",
        connectgaps=True
    ))

plane_avg = airplanes.groupby('Year', as_index=False)[('Fatalities', 'Ground')].mean()
print(plane_avg)

airplane_num = 0
sorted_plane_years = plane_avg['Year'].sort_values(ascending=True,inplace=False)
for year in sorted_plane_years:
    all_years.append(year)
    df = plane_avg[plane_avg['Year'] == year]
    airplane_num += 1
    fig.add_trace(go.Scatter(
        x=df['Fatalities'],
        y=df['Ground'],
        mode="lines",
        marker=dict(
            color="blue",
            showscale=True,
            symbol="circle"
        ),
        visible=False,
        name="Fatalities vs Ground",
        connectgaps=True
    ))

print(train_num)
print(airplane_num)

steps = []
for i in range(len(fig.data)):
    step = dict(
        method="update",
        label=str(all_years[i]),
        args=[{"visible": [False] * len(fig.data)}, #i think this is the problem bc it's setting the slider to fig length
              {"title": "Slider switched to Year: " + str(all_years[i])}],  # layout attribute
    )
    step["args"][0]["visible"][i] = True  # Toggle i'th trace to "visible"
    steps.append(step)

train_steps = []
for i, year in enumerate(train_avg['YEAR']):
    print(i, year)
    step = dict(
        method="update",
        label=str(all_years[i]),
        args=[{"visible": [False] * train_num}, 
              {"title": "Slider switched to Year: " + str(all_years[i])},
              {"x": [train_avg[train_avg['YEAR'] == year]['TRNSPD']]},
              {"y": [train_avg[train_avg['YEAR'] == year]['TONS']]}
              ],  # layout attribute
    )
    step["args"][0]["visible"][i] = True  # Toggle i'th trace to "visible"
    train_steps.append(step)

plane_steps = []
for i in range(airplane_num):
    step = dict(
        method="update",
        label=str(all_years[i + train_num]),
        args=[{"visible": [False] * airplane_num}, #i think this is the problem bc it's setting the slider to fig length
              {"title": "Slider switched to Year: " + str(all_years[i + train_num])}],  # layout attribute
    )
    step["args"][0]["visible"][i] = True  # Toggle i'th trace to "visible"
    plane_steps.append(step)

sliders = [dict(
    active=train_num,
    currentvalue={"prefix": "Year: "},
    pad={"t": 50},
    steps=train_steps,
    visible=True
), dict(
    active=airplane_num,
    currentvalue={"prefix": "Year: "},
    pad={"t": 50},
    steps=plane_steps,
    visible=False
)
]


print(train_steps)

print(len(fig.data))

fig.update_layout(
    template="plotly_white",
    xaxis_title="x here",
    title="something",
    sliders=sliders
)

button_height = 0.15
x1_loc = 0.00
y1_loc = 1.15

train_visible = [True] * train_num + [False] * airplane_num
airplane_visible = [False] * train_num + [True] * airplane_num
# DROPDOWN MENUS
fig.update_layout(
    # DEFINE A LIST OF THE VARIOUS BUTTONS (STORED AS DICTIONARIES)
    updatemenus=[
        # BUTTON-1: OPACITY
        dict(
            # NOTICE THAT THE OPTION ARE A LIST OF DICTIONARIES
            # IMPORTANT: args specifies the key-value pairing for what to change
            buttons=[
                dict(
                    label="Trains",            # LABEL SHOWN TO USER
                    method="update",                    # MODIFICATION TYPE (SEE ABOVE)
                     args=[{"visible": [True] * train_num + [False] * airplane_num},  # BOOLEAN VALUES FOR EACH TRACE
                           {"title": "Trains"},
                           {'sliders': sliders[0]}],
                     ),
                dict(
                    label="Airplanes",               # LABEL SHOWN TO USER
                    method="update",           # MODIFICATION TYPE (SEE ABOVE)
                     args=[{"visible": [False] * train_num + [True] * airplane_num},
                           {"title": "Airplanes"},
                           {'sliders': sliders[1]}],
                    execute=True
                     ),            
            ],
            # PLACEMENT AND META DATA FOR THE BUTTON
            direction="down",
            showactive=True,  # HIGHLIGHTS ACTIVE DROPDOWN ITEM OR ACTIVE BUTTON IF TRUE
            pad={"r": 10, "t": 10},  # PADDING
            x=x1_loc,  # POSITION
            y=y1_loc,
            xanchor="left",  # ANCHOR POINT
            yanchor="top",
        )
    ]
) 


# # SHOW
fig.show()
```
